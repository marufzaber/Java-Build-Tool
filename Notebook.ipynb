# Effect in # files

```python
file_ents_to_compl = pd.read_csv('data_log.csv', sep=',', encoding='utf-8')
```

### Effect by project

```python
pd.pivot_table(file_ents_to_compl,
               index='project_id',
               values=['file_based', 'entity_based'],
               aggfunc='median').describe()
# Difference between columns is difference in compilation efforts
```

```python
pd.pivot_table(file_ents_to_compl,
               index='file_id',
               values=['file_based', 'entity_based'],
               aggfunc='median').describe()
# Difference between columns is difference in compilation efforts
```
